```python
#Shelby Long
questions = [
  "Who was the first American woman in space? ", 
  "True or False: 5 kilometers == 5000 meters? ", 
  "(5+3)/2*10 =  ? ", 
  "Given the list [8, 'Orbit', 'Trajectory', 45], what entry is at index 2? ", 
  "What is the minimum crew size for the ISS? " 
] 
print()
correct_answers = ["Sally Ride", "true", "40", "Trajectory", "3"]
candidate_answers = []
candidate_name=input("What is your name?  ")
def ask_for_name():
  print(f"Hello {candidate_name}, welcome to the console!")

  return candidate_name
  


def ask_question():
  answer= 0
  index= 0
  print('\n')
  for question in range(len(questions)):
    answer = input(questions[index])
    candidate_answers.append(answer)
    index += 1
  
  print("___________________________________________________________\n")

  print(f'Candidate Name: {candidate_name}\n')
  for i in range(len(questions)):
    
    question_output = " {} {} {}"
    candidate_ans_output= "Your Anwswer: {}"
    correct_ans_output= "Correct Answer: {}"

    print(question_output.format(i+1, ':', questions[i]))
    print(candidate_ans_output.format(candidate_answers[i]))
    print(correct_ans_output.format(correct_answers[i]))
    print('\n')
          
  # print("___________________________________________________________\n")
  # print(f'Candidate Name: {candidate_name}\n')
  # print(f'Question 1: {questions[0]}\nYOUR ANSWER: {candidate_answers[0]}\nCORRECT ANSWER: {correct_answers[0]}')
  # print("\n")
  # print(f'Question 2: {questions[1]}\nYOUR ANSWER: {candidate_answers[1]}\nCORRECT ANSWER: {correct_answers[1]}')
  # print("\n")
  # print(f'Question 3: {questions[2]}\nYOUR ANSWER: {candidate_answers[2]}\nCORRECT ANSWER: {correct_answers[2]}')
  # print("\n")
  # print(f'Question 4: {questions[3]}\nYOUR ANSWER: {candidate_answers[3]}\nCORRECT ANSWER: {correct_answers[3]}')
  # print("\n")
  # print(f'Question 5: {questions[4]}\nYOUR ANSWER: {candidate_answers[4]}\nCORRECT ANSWER: {correct_answers[4]}')
  # print("\n")
  
  # return candidate_answers



def grade_quiz(candidate_answers):
  num_correct_answers = 0
  num_questions = 5
  
  
  for index in range(len(correct_answers)):
      if candidate_answers[index].lower() != correct_answers[index].lower():
         num_correct_answers = num_correct_answers 
         index += 1
      elif candidate_answers[index].lower() == correct_answers[index].lower():
        num_correct_answers += 1
        index += 1


  grade = (num_correct_answers)/(num_questions)*(100)
  print(f'>>> Overall Grade: {grade}% ({num_correct_answers} of {num_questions} responses correct) <<<')

  if grade>=80:
     print(">>> Status: PASSED <<<")
  else:
     print(">>> Status: FAILED <<<")
  
  return grade






def run_program():
  ask_for_name()
  ask_question()
  grade_quiz(candidate_answers)

run_program()

```
